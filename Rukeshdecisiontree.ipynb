```python
training_features, test_features, training_target, test_target = train_test_split(
    features, target, random_state=430)
```

**Explanation**


In the above line of code testing and the training data is separate. Training_features, training_target are the training dataset used to train the model. Test_features, test_target are the testing data to test the train model. Randam_state select the random data to train and testing the data. We have to split the dataset as traina nd test because in machine learning we have to give ability to machine to learn the new things from the dataset. The data we feed to the machine is used to learn the pattern and predict class. 

Test dataset helps to determine wheather the machine has the learn the pattern or not. Same data cannot be provide to the machine for testing as it has already seen the data. It will obiously predict the actual output. If the new dataset is given to predict then we can distinguist the result is right or not.

```python
print(training_features.shape, test_features.shape)
```

**Explantion**

Above line of code print the training and the testing data of label feature. The total data in the datast was (200,4). 200 rows of data with the 4 feature. The data is split in the ratio (75:25). 77% of the total data is training dataset and the remaining 25 is the testing dataset. 

```python
dtc = DecisionTreeClassifier(criterion='entropy')
dtc
```

```python
model = dtc.fit(training_features, training_target)
```

```python
predictions = model.predict(test_features)
```

```python
matrix = confusion_matrix(test_target, predictions)
```

**Explanation**

Confusion matrix is the table which is used to know the performance of the classification model for the testing data.  In the confusion matrix there are four terms 
Accuracy: How often is the classifier correct?

Misclassification Rate: how often is it wrong?

True Positive Rate: When it's actually yes, how often does it predict yes?

False Positive Rate: When it's actually no, how often does it predict yes?

True Negative Rate: When it's actually no, how often does it predict no?

Precision: When it predicts yes, how often is it correct?

Prevalence: How often does the yes condition actually occur in our sample?

In the above code matrix varaible is created which holds the performance of the classifier. Builtin funciton of the sklearn is used to make the confusion matrix. The testing data is passed as the parameter of the confusion matrix.


```python
print(matrix)
```

**Explanation**

Above line of code is used to print the confusion matrix of the classification. 



```python
print(classification_report(test_target, predictions))
```

**Explanation**

 Above line of code is used to print the report of the classification problem. There is two parameter to generate the report. One is correct target values and another is predited target values. 
 
Accuracy: (TP+TN)/total

Recall: TP/actual yes

Precision: TP/predicted yes

F1-score = (2 * Recall * Precision) / (Recall + Precision)

Micro Average = [ Sum (Correct Classification) / Total Samples ]

Weighted Average = [ Sum (Precision * Total Predicted ) ] / Total
Sample

Above line of code print the confusion matrix accuracy, macro avg and weighted avg. AVG stands for the average. Precision, Recall, F1-score, support are the parameter to measure the confusion matrix.
